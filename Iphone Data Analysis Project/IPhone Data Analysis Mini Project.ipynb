```python
# Using Regex I am creating a new column called "color" extracting from the Product Name using the extract function
df['color'] = df['Product Name'].str.extract(r'\(([^,]+)')
```

```python
# Using Regex I am creating a new column called "Storage" extracting from the Product Name using the extract function
df['Storage'] = df['Product Name'].str.extract(r',([^)]*)')
```
